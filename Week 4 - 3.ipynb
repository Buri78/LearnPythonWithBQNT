** Retrieve multiple data fields with BQL **

*Link for BQL Basics Webinar Video:*

```python
import bql
bq = bql.Service()
```

```python
name = bq.data.name()
px = bq.data.px_last()
eps = bq.data.is_eps()

index = bq.univ.members('INDU Index')

req = bql.Request(index, [name, px, eps])
res = bq.execute(req)
```

** Retrieve data field by field **

```python
data_name = res[0].df()
data_px = res[1].df()
data_eps = res[2].df()

data_eps.head()
```

** Concatenate all DataFrames together **

** Rename the column headers **

```python
df = df.rename(columns={'NAME()': 'NAME', 'PX_LAST()': 'PRICE', 'IS_EPS()': 'EPS'})
df.head()
```

** Column wise calculations **

*Calculate P/E Ratio using PRICE and EPS columns*

```python
df['PE'] = df['PRICE'] / df['EPS']
df.head()
```

** Advanced calculation based on column data with Lambda **

*Remove the negative PE*

```python
import numpy as np

df['PE'] = df['PE'].apply(lambda x: np.NaN if x < 0 else x)
df.head()
```
